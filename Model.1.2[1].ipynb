```python
# Base directory containing both folders
base_directory = 'wav_audio_files'

# Paths to the two folders
folder_husary = os.path.join(base_directory, 'husary')
folder_alafasy = os.path.join(base_directory, 'alafasy')

# Function to print paths of all files and directories in a given folder
def print_directory_contents(folder_path, folder_name):
    print(f"Path of {folder_name} folder: {folder_path}")
    for root, dirs, files in os.walk(folder_path):
        # Print directories
        for dir_name in dirs:
            print(f"Directory: {os.path.join(root, dir_name)}")
        # Print files
        for file_name in files:
            print(f"File: {os.path.join(root, file_name)}")
    print()  # Print a newline for better readability

# Print contents of husary folder
print_directory_contents(folder_husary, 'husary')

# Print contents of alafasy folder
print_directory_contents(folder_alafasy, 'alafasy')
```

```python
def load_and_print_audio_details(audio_files, folder_name):
    audio_data_list = []
    sr_list = []
    for audio_file in audio_files:
        try:
            # Load the audio file
            audio_data, sr = librosa.load(audio_file, sr=None, mono=True)
            
            audio_data_list.append(audio_data)
            sr_list.append(sr)
            
            # Print details
            print(f'Loaded Audio from {folder_name}: {audio_file}')
            print(f'Original Sample Rate: {sr} Hz')
            print(f'Original Number of Samples: {len(audio_data)}')
            print()
        except Exception as e:
            print(f"Error loading audio file {audio_file}: {e}")
    
    return audio_data_list, sr_list

# Retrieve audio files
audio_files_alafasy = list_audio_files(folder_alafasy)
audio_files_husary = list_audio_files(folder_husary)

# Check if there are audio files in both folders and process them
if audio_files_alafasy and audio_files_husary:
    # Process and print details for all audio files in the alafasy folder
    alafasy_data, alafasy_sr = load_and_print_audio_details(audio_files_alafasy, 'alafasy')
    
    # Process and print details for all audio files in the husary folder
    husary_data, husary_sr = load_and_print_audio_details(audio_files_husary, 'husary')
else:
    raise FileNotFoundError("No audio files found in one or both directories.")


```

```python
output_folder_alafasy = 'segmented_audio_files/alafasy_normalized'
output_folder_husary = 'segmented_audio_files/husary_normalized'
os.makedirs(output_folder_alafasy, exist_ok=True)
os.makedirs(output_folder_husary, exist_ok=True)

def list_audio_files(directory):
    audio_files = []
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.wav'):
                audio_files.append(os.path.join(root, file))
    return audio_files

# Function to normalize and save audio files
def normalize_and_save_audio_files(audio_files, output_folder):
    if not audio_files:
        print(f"No audio files found in the folder {output_folder}.")
        return

    for audio_file in audio_files:
        if not os.path.isfile(audio_file):
            print(f"File not found: {audio_file}")
            continue
        
        try:
            # Load audio file
            audio_data, sr = librosa.load(audio_file, sr=None, mono=True)
            
            # Normalize audio
            max_volume = np.max(np.abs(audio_data))
            if max_volume > 0:
                audio_normalized = audio_data / max_volume
            else:
                audio_normalized = audio_data
            
            # Create output path
            relative_path = os.path.relpath(audio_file, base_directory)
            output_path = os.path.join(output_folder, relative_path)
            os.makedirs(os.path.dirname(output_path), exist_ok=True)
            
            # Save normalized audio
            sf.write(output_path, audio_normalized, sr)
            
            # Print success message
            print(f'Successfully normalized and saved: {output_path}')
        except Exception as e:
            print(f"Error processing {audio_file}: {e}")

# Retrieve audio files
audio_files_alafasy = list_audio_files(folder_alafasy)
audio_files_husary = list_audio_files(folder_husary)


print(f"Audio files in alafasy folder: {audio_files_alafasy}")
print(f"Audio files in husary folder: {audio_files_husary}")

# Normalize and save all audio files
normalize_and_save_audio_files(audio_files_alafasy, output_folder_alafasy)
normalize_and_save_audio_files(audio_files_husary, output_folder_husary)
```

```python
frame_length = 2048
hop_length = 512

# Extract MFCC features for the first audio file in each folder
if alafasy_data and alafasy_sr and husary_data and husary_sr:
    Ref_mfcc_features = librosa.feature.mfcc(y=alafasy_data[0], sr=alafasy_sr[0], n_mfcc=13, n_fft=frame_length, hop_length=hop_length)
    User_mfcc_features = librosa.feature.mfcc(y=husary_data[0], sr=husary_sr[0], n_mfcc=13, n_fft=frame_length, hop_length=hop_length)
    
    print("MFCC Features Shape (Ref):", Ref_mfcc_features.shape)
    print("MFCC Features Shape (User):", User_mfcc_features.shape)
else:
    print("Audio data or sample rate missing for MFCC extraction.")
```

```python
# Resample User_mfcc_features to match the number of frames in Ref_mfcc_features NOM OF Column
def resample_features(features, target_length):
    current_length = features.shape[1]
    x = np.arange(current_length)
    x_new = np.linspace(0, current_length-1, target_length)
    interpolator = interp1d(x, features, kind='linear', axis=1, fill_value='extrapolate')
    return interpolator(x_new)

resampled_User_mfcc_features = resample_features(User_mfcc_features, Ref_mfcc_features.shape[1])
scaler = StandardScaler()
Ref_mfcc_features_scaled = scaler.fit_transform(Ref_mfcc_features.T).T
User_mfcc_features_scaled = scaler.transform(resampled_User_mfcc_features.T).T

print(Ref_mfcc_features_scaled.shape)
print(User_mfcc_features_scaled.shape)
```

#  Add new dimension


```python
import numpy as np

def generate_data(ref_features, user_features):
    # labels 0 for reference , 1 for user audio
    ref_labels = np.zeros(ref_features.shape[0])
    user_labels = np.ones(user_features.shape[0])
    
    # Combine reference and user features
    X = np.concatenate([ref_features, user_features], axis=0)
    y = np.concatenate([ref_labels, user_labels], axis=0)
    
    # Shuffle data
    indices = np.arange(X.shape[0])
    np.random.shuffle(indices)
    X = X[indices]
    y = y[indices]
    
    return X, y

X_train, y_train = generate_data(Ref_mfcc_features_scaled, User_mfcc_features_scaled)

print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of y_train: {y_train}")
```

```python

model = Sequential()

# Add LSTM layers
model.add(LSTM(128, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(LSTM(64, return_sequences=True))
model.add(LSTM(32, return_sequences=True))
model.add(LSTM(16, return_sequences=False))

model.add(Dropout(0.5))


model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))

model.add(Dense(1, activation='sigmoid'))


model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
# Train the model
history = model.fit(
    X_train,        
    y_train,        # Training labels
    epochs=15,      
    batch_size=15,  
    validation_split=0.2  # Fraction of data to be used as validation set
)
```

```python
X_test, y_test = generate_data(Ref_mfcc_features_scaled[:100], User_mfcc_features_scaled[:100])
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy}")
print(f"Test Loss: {loss}")
```
